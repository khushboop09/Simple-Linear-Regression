```python
#import the libraries
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

#read a random dataset
dataf = pd.read_csv('\dataset.txt')
x = dataf[[0]]
y = dataf[[1]]

#train the model
reg = linear_model.LinearRegression()
reg.fit(x,y)

#Error and variance score
mean_error = np.mean((reg.predict(x) - y) ** 2)
print 'Mean Squared Error: %.3f'%mean_error
variance_score = reg.score(x, y)
print 'Variance score: %.3f' %variance_score 

#plot
plt.scatter(x,y)
plt.plot(x,reg.predict(x))
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Some random dataset')
plt.show()
```
